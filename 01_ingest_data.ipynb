```python
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")
print("Bronze database ready")
```

```python
spark.sql("DROP TABLE IF EXISTS bronze.raw_customers")
print("Table dropped")
```

```python
# Read customers CSV from volume
df_customers = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/main/default/raw_data/customers (1).csv")

# Save as Delta table
df_customers.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.raw_customers")

print(f"Saved {df_customers.count()} customers")
display(df_customers)
```

```python
df_orders = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/main/default/raw_data/orders (1).csv")

df_orders.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.raw_orders")

print(f"Saved {df_orders.count()} orders")
display(df_orders)
```

```python
df_events = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/main/default/raw_data/email_events.csv")

df_events.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.raw_email_events")

print(f"Saved {df_events.count()} email events")
display(df_events)
```
